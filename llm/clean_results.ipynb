```python
import pandas as pd
```

```python
df = pd.read_csv('./datasets/results_26_27.csv')
```

```python
def remove_earthquake_prefix(row):
    text = row.get("finetuned_26_27", "")
    location = row.get("text", "")
    
    # Ensure both are strings and not NaN
    if pd.isna(text) or pd.isna(location):
        return text  # or "" if you prefer empty string on bad rows
    
    text = str(text)
    location = str(location)

    to_remove = f"Identify locations like country, city, or any local areas in the following sentence and write them in a comma-separated manner (no additional details required):, {location}"
    return text.replace(to_remove, "").strip()

df["finetuned_26_27"] = df.apply(remove_earthquake_prefix, axis=1)
```

```python
df.to_csv('./datasets/results_26_27.csv', index=False)
```
